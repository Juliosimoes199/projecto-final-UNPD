# Tratando as Colunas categÃ³ricas

```python
def selecionar_colunas_uteis(df, colunas_categoricas, coluna_alvo, tipo_alvo):
    """
    Seleciona colunas categÃ³ricas que tÃªm uma relaÃ§Ã£o estatisticamente
    significativa com a variÃ¡vel alvo.
    
    Args:
        df (DataFrame): O DataFrame a ser analisado.
        colunas_categoricas (list): Lista de nomes de colunas categÃ³ricas.
        coluna_alvo (str): Nome da coluna alvo.
        tipo_alvo (str): 'numerico' ou 'categorico'.
        
    Returns:
        list: Uma lista com os nomes das colunas que devem ser mantidas.
    """
    colunas_a_manter = []
    
    # Loop atravÃ©s de cada coluna categÃ³rica
    for col in colunas_categoricas:
        if tipo_alvo == 'numerico':
            # --- Se a variÃ¡vel alvo for NUMÃ‰RICA (ex: PreÃ§o) ---
            # Agrupa os dados numÃ©ricos por categoria
            grupos = [df[df[col] == categoria][coluna_alvo] for categoria in df[col].unique()]
            
            # Realiza o teste de ANOVA
            f_estatistica, p_valor = f_oneway(*grupos)
            
        elif tipo_alvo == 'categorico':
            # --- Se a variÃ¡vel alvo for CATEGÃ“RICA (ex: Churn) ---
            # Cria a tabela de contingÃªncia
            tabela_contingencia = pd.crosstab(df[col], df[coluna_alvo])
            
            # Realiza o teste Qui-quadrado
            chi2, p_valor, _, _ = chi2_contingency(tabela_contingencia)
            
        else:
            raise ValueError("O tipo_alvo deve ser 'numerico' ou 'categorico'.")
            
        # Verifica o p-valor para decidir se a coluna Ã© Ãºtil
        if p_valor < 0.05:
            print(f"Mantendo '{col}': p-valor = {p_valor:.4f} (significativo)")
            colunas_a_manter.append(col)
        else:
            print(f"Descartando '{col}': p-valor = {p_valor:.4f} (nÃ£o significativo)")
    
    return colunas_a_manter

# Exemplo de uso:
# Supondo que 'df' Ã© o seu DataFrame, 'SalePrice' Ã© a variÃ¡vel alvo numÃ©rica
# e vocÃª tem uma lista de colunas categÃ³ricas para testar
colunas_uteis = selecionar_colunas_uteis(df, obj, 'is_canceled', 'numerico')

print("\nColunas categÃ³ricas a serem usadas no modelo:", colunas_uteis)
```

```python
## Stepwise Selection for OLS Regression ##
def ols_stepwise_selection(x_data,
                           y_data, 
                           initial_list  = [], 
                           threshold_in  = 0.01, 
                           threshold_out = 0.05, 
                           verbose = True):
    """
    Executa uma seleÃ§Ã£o de recursos para frente e para trÃ¡s com base nos valores-p de statsmodels.api.OLS.

Argumentos:
x_data : pandas.DataFrame
DataFrame com recursos candidatos.
y_data : array-like
A variÃ¡vel de destino.
initial_list : list
Lista de recursos (nomes de colunas de x) para comeÃ§ar.
threshold_in : float
Inclui um recurso se seu valor-p < threshold_in.
threshold_out : float
Exclui um recurso se seu valor-p > threshold_out.
verbose : bool
Se deve ou nÃ£o imprimir a sequÃªncia de inclusÃµes e exclusÃµes.

Retorna:
list
A lista de recursos selecionados.

ObservaÃ§Ã£o: Sempre defina threshold_in < threshold_out para evitar loops infinitos.
    """

    # required imports
    import statsmodels.api as sm
    import pandas as pd


    # definindo lista de placeholer
    included = list(initial_list)


    # loop sobre cada caracterÃ­stica x atÃ© que nÃ£o haja mais valores p significativos
    while True:
        changed = False

        # passo Ã  frente: adicionando um recurso x
        excluded = [col for col in x_data.columns if col not in included]
        new_pvals = pd.Series(dtype = float, index = excluded)


        # modelo de ajuste com recurso candidato adicional
        for new_column in excluded:

            model = sm.OLS(y_data,
                           sm.add_constant(x_data[included + [new_column]])).fit()

            new_pvals[new_column] = model.pvalues[new_column]


        if not new_pvals.empty:
            best_pval = new_pvals.min()
            if best_pval < threshold_in:
                best_feature = new_pvals.idxmin()  # Use idxmin() em vez de argmin()
                included.append(best_feature)
                changed = True
                if verbose:
                    print('Add  {:30} with p-value {:.6}'.format(best_feature, best_pval))


        # passo para trÃ¡s: potencialmente removendo um recurso x
        if included:
            model = sm.OLS(y_data, sm.add_constant(x_data[included])).fit()

            # excluindo o valor p de interceptaÃ§Ã£o (primeiro elemento)
            pvals = model.pvalues.iloc[1:]

            # garantindo que o modelo nÃ£o esteja vazio
            if not pvals.empty:
                worst_pval = pvals.max()
                if worst_pval > threshold_out:
                    worst_feature = pvals.idxmax()  # Usando idxmax() em vez de argmax()
                    included.remove(worst_feature)
                    changed = True
                    if verbose:
                        print('Drop {:30} with p-value {:.6}'.format(worst_feature, worst_pval))


        # parando o loop se otimizado
        if not changed:
            break


    # retornando os recursos x do modelo passo a passo
    return included
```

# Feature Selection

```python
import numpy as np
from sklearn.linear_model import ElasticNet
from sklearn.model_selection import GridSearchCV
from sklearn.feature_selection import SelectFromModel
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder

# Suponha que 'preprocessor' Ã© o seu prÃ©-processador jÃ¡ definido
# E X_train, y_train sÃ£o os seus dados de treino
X = df_concat.drop(["is_canceled"], axis=1)
y = df_concat.is_canceled

numeric_features = X.columns

# 2. Crie o prÃ©-processador (esta Ã© a variÃ¡vel que estava faltando)
# Ele aplica o StandardScaler nas colunas numÃ©ricas e o OneHotEncoder nas categÃ³ricas
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numeric_features)
    ])


# Defina a pipeline do modelo
pipeline = Pipeline([
    ('preprocessor', preprocessor), # Seu prÃ©-processador
    ('elasticnet', ElasticNet(max_iter=5000))
])

# Defina o espaÃ§o de busca de hiperparÃ¢metros
param_grid = {
    'elasticnet__alpha': np.logspace(-4, 0, 10),  # Exemplo: 10 valores entre 0.0001 e 1
    'elasticnet__l1_ratio': np.linspace(0.1, 1.0, 10) # Exemplo: 10 valores entre 0.1 e 1.0
}

# Configure e execute o GridSearchCV
grid_search = GridSearchCV(
    pipeline,
    param_grid,
    cv=5,                 # NÃºmero de folds para validaÃ§Ã£o cruzada
    scoring='r2',         # MÃ©trica de avaliaÃ§Ã£o (pode ser 'neg_mean_squared_error' tambÃ©m)
    n_jobs=-1,            # Use todos os nÃºcleos do processador
    verbose=1
)

grid_search.fit(X, y)

# Exiba os melhores parÃ¢metros e a melhor pontuaÃ§Ã£o
print("Melhores parÃ¢metros encontrados: ", grid_search.best_params_)
print("Melhor pontuaÃ§Ã£o RÂ²: ", grid_search.best_score_)

# O melhor modelo Ã© acessado por grid_search.best_estimator_
best_model_grid = grid_search.best_estimator_
```

```python

# 1. Extraia o estimador ElasticNet da pipeline
#    O nome 'elasticnet' Ã© o que vocÃª definiu na sua pipeline.
elasticnet_estimator = best_model_grid.named_steps["elasticnet"]

# 2. Extraia o prÃ©-processador da pipeline
preprocessor = best_model_grid.named_steps["preprocessor"]

# 3. Transforme os dados de entrada usando o prÃ©-processador
X_transformed = preprocessor.transform(X)

# 4. Use o SelectFromModel com o estimador ElasticNet e os dados transformados
seletor = SelectFromModel(estimator=elasticnet_estimator)
seletor.fit(X_transformed, y)

# 5. Obtenha as features selecionadas
feat_mask = seletor.get_support()

# 6. Identifique os nomes das colunas originais selecionadas
#    Aqui Ã© onde vocÃª precisa de mais cuidado, pois a ordem das colunas
#    pode ser alterada pelo prÃ©-processador.
#    VocÃª precisaria de uma forma de mapear as features transformadas de volta para
#    os nomes originais das colunas.

# Uma forma simplificada para datasets com apenas colunas numÃ©ricas:
feat = X.columns[feat_mask]
X_grid = X[feat]
X_grid

Xtrain, Xtest, ytrain, ytest = train_test_split(X_grid, y, test_size=0.2, random_state=42)
# Para um prÃ©-processador com diferentes transformaÃ§Ãµes, a lÃ³gica seria mais complexa.
```

```python
from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.metrics import precision_score, make_scorer
```

```python
# Supondo que vocÃª tem seus dados:
# X = ... (Todas as features)
# y = ... (VariÃ¡vel alvo)

# Definir a mÃ©trica de scoring (ex: Precision, que vocÃª estava a usar)
# Ã‰ importante usar a mÃ©trica de interesse para otimizar
scorer = make_scorer(recall_score, average='macro') # Use 'macro' ou o 'average' adequado para o seu problema
```

```python
# 1. Definir o dicionÃ¡rio de hiperparÃ¢metros para testar
param_grid = {
    # Reduzir a profundidade das Ã¡rvores
    'max_depth': [3, 4, 5], 
    
    # Reduzir a taxa de aprendizado para que o modelo aprenda mais devagar
    'learning_rate': [0.01, 0.05, 0.1],
    
    # Adicionar regularizaÃ§Ã£o L2
    'reg_lambda': [1, 5, 10], 
    
    # Subamostragem de colunas para cada Ã¡rvore (ajuda a evitar overfitting)
    'colsample_bytree': [0.7, 0.9], 
}
```
